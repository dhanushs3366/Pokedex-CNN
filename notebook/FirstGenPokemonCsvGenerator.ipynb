```python
def create_df_from_dataset(dataset_dir,csv_ref_path):
    pokemon_names=os.listdir(dataset_dir)
    pokemon_df_details=[]
    for pokemon_name in pokemon_names: #pokemon_names->[Abra,...], pokemon_name->Abra
        pokemon_name_path=os.path.join(dataset_dir,pokemon_name)
        pokemon_img_paths=os.listdir(pokemon_name_path)
        for pokemon_img in pokemon_img_paths:
            pokemon_img_path=os.path.join(pokemon_name_path,pokemon_img)
            pokemon_df=pd.read_csv(csv_ref_path)
            pokemon_row=pokemon_df[(pokemon_df[' Name']==pokemon_name)]
            pokemon_detail={
                "name":pokemon_row[' Name'].iloc[0],
                "id":pokemon_row['Number'].iloc[0],
                "img":pokemon_img_path,
                "type1":pokemon_row[' Type1'].iloc[0]
            }
            if pokemon_row[' Types'].iloc[0]==2:
                pokemon_detail["type2"]=pokemon_row[' Type2'].iloc[0]
            pokemon_df_details.append(pokemon_detail)
            
    return pokemon_df_details
```

```python
def create_df_from_dataset(dataset_dir, csv_ref_path):
    pokemon_names = os.listdir(dataset_dir)
    pokemon_df_details = []

    for pokemon_name in pokemon_names:
        pokemon_name_path = os.path.join(dataset_dir, pokemon_name)
        pokemon_img_paths = os.listdir(pokemon_name_path)

        for pokemon_img in pokemon_img_paths:
            pokemon_img_path = os.path.join(pokemon_name_path, pokemon_img)
            pokemon_df = pd.read_csv(csv_ref_path)
            pokemon_row = pokemon_df[(pokemon_df['Name'] == pokemon_name)]

            if not pokemon_row.empty:
                pokemon_detail = {
                    "name": pokemon_row['Name'].iloc[0],
                    "id": pokemon_row['#'].iloc[0],
                    "image_path": pokemon_img_path,
                    "type1": pokemon_row['Type 1'].iloc[0]
                }

                if not pd.isna(pokemon_row['Type 2'].iloc[0]):
                    pokemon_detail["type2"] = pokemon_row['Type 2'].iloc[0]

                pokemon_df_details.append(pokemon_detail)
            else:
                print(f"No data found for {pokemon_name} in the CSV.")

    return pokemon_df_details
```

```python
pokemon_details=create_df_from_dataset(dataset_dir,pokemon_csv_ref_path)
```

```python
pokemon_found=len(pokemon_details)/4
total_pokemon=len(os.listdir(dataset_dir))
print(f"There are {pokemon_found} and {total_pokemon-pokemon_found} pokemon not found")
```

```python
pokemon_df=pd.DataFrame(pokemon_details)
```
